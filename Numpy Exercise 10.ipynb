**Using Numpy, create an array 10x10 filled with number 255 and set the data type to float. Print this array to the console as shown below.**

```python
import numpy as np
```

```python
np.ones((10,10), dtype='float')*255
```

```python
print(np.ones((10,10), dtype='float')*255)
```

```python
np.full((10,10), fill_value=255, dtype='float')
```

```python
print(np.full((10,10), fill_value=255, dtype='float'))
```
