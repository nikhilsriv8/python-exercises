**Random seed is set to 42. The following array is given:**



image = np.random.randint(
    low=0, high=256, size=(10, 10, 3), dtype=np.uint8
)


**Extract the first channel (index 0) of this image array and print it to the console.**



```python
import numpy as np


np.random.seed(42)

image = np.random.randint(
    low=0, high=256, size=(10, 10, 3), dtype=np.uint8
)
```

```python
image
```

```python
image[:, :, 0]
```

```python
print(image[:, :, 0])
```
